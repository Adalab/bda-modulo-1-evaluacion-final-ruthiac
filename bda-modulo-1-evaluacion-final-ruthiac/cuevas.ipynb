# FUNCIONES

 ðŸ“Œ**1) AGREGAR PRODUCTO**


```python
def agregar_producto (nombre, cantidad, precio):
    # Recorro los productos en el diccionario alimentos_perros y verifico si el producto ya existe
    for alimentos_perro in alimentos_perros:
        if alimentos_perro['nombre'] == nombre:
            print(f"El articulo {nombre} ya existe. Se actualizarÃ¡ la cantidad.")
            alimentos_perro['cantidad'] += cantidad
            print(f"Nueva cantidad de {nombre}: {alimentos_perro['cantidad']}")
            #si exite el alimento de perro no se agrega el precio
            print(f"Precio de {nombre}: {alimentos_perro['precio']} â‚¬, no se modifica porque ya existe.")
            #se encontro el articulo por eso se procede con un return para cortar la ejecucion
            return
    #si no se encontro el articulo, lo agrego como nuevo
    alimentos_perros.append({'nombre': nombre, 'precio': precio, 'cantidad': cantidad})


        
while True:
    articulo_seleccionado= input("Ingrese el nombre del articulo que desea agregar o actualizar: ")
    precio_nuevo_producto = float(input("Ingrese el precio del nuevo producto: "))  
    cantidad_nuevo_producto = int(input("Ingrese la cantidad del nuevo producto: "))      
    agregar_producto(articulo_seleccionado, cantidad_nuevo_producto, precio_nuevo_producto)

    continuar = input("Â¿Desea agregar o actualizar otro producto? (si/no): ")
    if continuar.lower() != 'si':
        break
    
```

 ðŸ“Œ**2)VER INVENTARIO** 

```python
def ver_inventario():
    print("Inventario actual:")
    for alimento in alimentos_perros:
        print(f"Articulo: {alimento['nombre']}, Precio: {alimento['precio']} â‚¬, Cantidad: {alimento['cantidad']}")
ver_inventario()
```

 ðŸ“Œ**3)BUSCAR PRODUCTO**

```python
buscar_prod=input("Ingrese el nombre del producto que desea buscar: ")  
def buscar_producto(nombre):
    for alimento in alimentos_perros:
        if alimento['nombre'].lower() == nombre.lower():
            print(f"Nombre: {alimento['nombre']}, Precio: {alimento['precio']} â‚¬, Cantidad: {alimento['cantidad']}")
            return   
    print(f"El producto '{nombre}' no se encontrÃ³ en el inventario.")
    
buscar_producto(buscar_prod)
```

ðŸ“Œ**4)ACTUALIZAR STOCK**

```python

stock= input("Ingrese el nombre del producto cuyo stock desea actualizar: ")

cantidad_stock= int(input("Si desea quitar al stock coloque un signo negativo adelante del numero y si desea agregar al stock agrege el numero: "))

def actualizar_stock(nombre, nueva_cantidad):
    for alimento in alimentos_perros:
        if alimento['nombre'].lower() == nombre.lower():
            alimento['cantidad'] = alimento['cantidad'] + nueva_cantidad
            print(f"Stock actualizado para {nombre}. Nueva cantidad: {alimento['cantidad']}")
            return
    print(f"El producto '{nombre}' no se encontrÃ³ en el inventario.")


actualizar_stock(stock, cantidad_stock)
```

ðŸ“Œ**5)ELIMINAR PRODUCTO**

```python
eli_producto=input("Ingrese el nombre del producto que desea eliminar del inventario: ")

def eliminar_producto(nombre):    
    for alimento in alimentos_perros:
        if alimento['nombre'].lower() == nombre.lower():
            alimentos_perros.remove(alimento)
            print(f"El producto '{nombre}' ha sido eliminado del inventario.")
            return
    print(f"El producto '{nombre}' no se encontrÃ³ en el inventario.")

eliminar_producto(eli_producto)
```

ðŸ“Œ**6)CALCULAR EL VALOR DEL INVENTARIO**

```python
def calcular_valor_inventario():
    valor_total = 0.0
    for alimento in alimentos_perros:
        valor_total += alimento['precio'] * alimento['cantidad']
    return valor_total

print(f"Valor total del inventario: {calcular_valor_inventario()} â‚¬")    
```

ðŸ“Œ**7)REALIZAR COMPRA**

```python
def realizar_compra():
    for cliente in clientes.values():
        carrito = []          # Lista de productos comprados en esta sesiÃ³n
        total_compra = 0.0    # Total de la compra del cliente
        quiere_comprar = True
        maximo_intentos = 3
        intentos = 0
        print(f"\nHola, {cliente['nombre']}, Â¡bienvenido! Â¿QuÃ© desea comprar?")

        while quiere_comprar and intentos < maximo_intentos:
            ver_inventario()
            nombre_producto = input("Ingrese el nombre del producto que desea comprar: ")
            
            # Verificar si el producto existe en el inventario
            existe = any(alimento['nombre'].lower() == nombre_producto.lower() for alimento in alimentos_perros)
            if not existe:
                intentos += 1
                print(f"Producto '{nombre_producto}' no encontrado. Intento {intentos} de {maximo_intentos}.\n")
                continue  # Volver al inicio del bucle

            # Buscar el producto y solicitar cantidad
            for alimento in alimentos_perros:
                if alimento['nombre'].lower().strip() == nombre_producto.lower().strip():
                    cantidad_deseada = int(input(f"Ingrese la cantidad que desea comprar (Disponibles: {alimento['cantidad']}): "))

                    if cantidad_deseada > alimento['cantidad']:
                        print(f"No hay suficiente stock de {nombre_producto}. Stock disponible: {alimento['cantidad']}\n")
                    else:
                        # Agregar al carrito y actualizar inventario y total
                        carrito.append({
                            'nombre': nombre_producto,
                            'cantidad': cantidad_deseada,
                            'precio': alimento['precio']
                        })
                        total_compra = total_compra + alimento['precio'] * cantidad_deseada
                        alimento['cantidad'] = alimento['cantidad'] - cantidad_deseada
                    break

            # Preguntar si quiere seguir comprando
            seguir = input("Â¿Desea comprar otro producto? (si/no): ").strip().lower()
            if seguir != 'si':
                quiere_comprar = False

        # Guardar la compra en el historial del cliente
        if carrito != []:
            cliente['compra'].extend(carrito) #agrega los productos del carrito a la lista de compras del cliente 
            print(f"\nCompra finalizada para {cliente['nombre']}. Total de la compra: {total_compra} â‚¬\n")
        else:
            print(f"{cliente['nombre']} no realizÃ³ ninguna compra.\n")
            # el cliente no realizo ninguna compra
realizar_compra()            
```
